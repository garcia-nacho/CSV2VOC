```
#Yolo 2 voc https://gist.github.com/goodhamgupta/7ca514458d24af980669b8b1c8bcdafd

# Script to convert yolo annotations to voc format

# Sample format
# <annotation>
#     <folder>_image_fashion</folder>
#     <filename>brooke-cagle-39574.jpg</filename>
#     <size>
#         <width>1200</width>
#         <height>800</height>
#         <depth>3</depth>
#     </size>
#     <segmented>0</segmented>
#     <object>
#         <name>head</name>
#         <pose>Unspecified</pose>
#         <truncated>0</truncated>
#         <difficult>0</difficult>
#         <bndbox>
#             <xmin>549</xmin>
#             <ymin>251</ymin>
#             <xmax>625</xmax>
#             <ymax>335</ymax>
#         </bndbox>
#     </object>
# <annotation>
import os
import xml.etree.cElementTree as ET
from PIL import Image
import csv

ANNOTATIONS_DIR_PREFIX = "/content/gdrive/My Drive/keras-yolo3-master/Pics/Pics/"

DESTINATION_DIR = "/content/gdrive/My Drive/keras-yolo3-master/Pics/annotsXML/"

CLASS_MAPPING = {'0': 'cell'}

def start(dir_name):
    if not os.path.exists(DESTINATION_DIR):
        os.makedirs(DESTINATION_DIR)
    for filename in os.listdir(ANNOTATIONS_DIR_PREFIX):
        if filename.endswith('csv'):
          read_file(filename)
        else:
          print("Skipping file: {}".format(filename))
            
def read_file(file_path):
    file_path2= ANNOTATIONS_DIR_PREFIX + file_path
    file_prefix = file_path.split(".csv")[0]

    with open(file_path2, 'r') as file:
        lines = csv.reader(file)
        voc_labels = []
                
        for line in lines:
          #print(line[0])
          voc = []
          voc.append(CLASS_MAPPING.get(line[0]))
          w=float(line[5])
          h=float(line[6])
          bbox_width = float(line[3]) * w
          bbox_height = float(line[4]) * h
          center_x = float(line[1]) * w
          center_y = float(line[2]) * h
          voc.append(center_x - (bbox_width / 2))
          voc.append(center_y - (bbox_height / 2))
          voc.append(center_x + (bbox_width / 2))
          voc.append(center_y + (bbox_height / 2))
          voc_labels.append(voc)
          #print(voc)
        create_file(file_prefix, w, h, voc_labels)
        print("Processing complete for file: {}".format(file_path))     
      
#read_file("/content/gdrive/My Drive/keras-yolo3-master/Pics/Pics/T4TimeCourseNoise.lsm - TimeCourseNoise #1.csv")       
def create_file(file_prefix, width, height, voc_labels):
    root = create_root(file_prefix, width, height)
    root = create_object_annotation(root, voc_labels)
    tree = ET.ElementTree(root)
    tree.write("{}/{}.xml".format(DESTINATION_DIR, file_prefix))
    
    
def create_object_annotation(root, voc_labels):
    for voc_label in voc_labels:
        obj = ET.SubElement(root, "object")
        ET.SubElement(obj, "name").text = voc_label[0]
        ET.SubElement(obj, "pose").text = "Unspecified"
        ET.SubElement(obj, "truncated").text = str(0)
        ET.SubElement(obj, "difficult").text = str(0)
        bbox = ET.SubElement(obj, "bndbox")
        ET.SubElement(bbox, "xmin").text = str(voc_label[1])
        ET.SubElement(bbox, "ymin").text = str(voc_label[2])
        ET.SubElement(bbox, "xmax").text = str(voc_label[3])
        ET.SubElement(bbox, "ymax").text = str(voc_label[4])
    return root

def create_root(file_prefix, width, height):
    root = ET.Element("annotations")
    ET.SubElement(root, "filename").text = "{}.jpg".format(file_prefix)
    ET.SubElement(root, "folder").text = "images"
    size = ET.SubElement(root, "size")
    ET.SubElement(size, "width").text = str(width)
    ET.SubElement(size, "height").text = str(height)
    ET.SubElement(size, "depth").text = "3"
    return root
  
def start(dir_name):
    if not os.path.exists(DESTINATION_DIR):
        os.makedirs(DESTINATION_DIR)
    for filename in os.listdir(ANNOTATIONS_DIR_PREFIX):
        if filename.endswith('csv'):
            read_file(filename)
            print("Reading File")
        else:
            print("Skipping file: {}".format(filename))  
```
